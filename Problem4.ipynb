# Largest Palindrome Product

```python
import numpy as np
```

```python
start = 1
stop = 999

N = (stop - start + 1)

M = np.zeros((N,N))

num = start
for i in range(N):
  for j in range(N):
    M[i, j] = (i + 1) * (j + 1)
```

```python
allnums = M[99:, 99:].reshape(-1, 1)
```

```python
i = 1
trial = int(allnums[-i, 0])
palindromes = []

for i in range(len(allnums)):
  if str(int(allnums[i, 0]))[:3] == (str(int(allnums[i, 0]))[-3:][::-1]):
    palindromes.append(int(allnums[i, 0]))
```

```python
max(palindromes)
```
