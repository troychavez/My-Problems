# Smallest Multiple

What is the smallest postive number that is evenly divisible by all numbers from 1 to 20? 

```python
i = 1
for k in (range(1, 21)):
    if i % k > 0:
        for j in range(1, 21):
            if (i*j) % k == 0:
                i *= j
                break
```
