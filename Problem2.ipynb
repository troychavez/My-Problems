# Even Fibonacci Numbers

By considering the terms in the Fibonacci sequence whose values do not exceed four million, find the sum of the even-valued terms.

Recursion is a method of solving a computational problem whose solution depends on the solution of smaller instances of that problem:

```python
# A recursive function might be defined as follows:

def fib_efficient(n, d):
  if n in d:
    return d[n] # For the first values of the Fibonacci sequence
  else:
    ans = fib_efficient(n - 1, d) + fib_efficient(n - 2, d) # F_n = F_(n-1) + F_(n-2)
    d[n] = ans
    return ans
```

```python
# In Python, we can create a dictionary (with the first numbers)
# and fill it ths subsequent Fibonacci values that do exceed four million:

d = {1:1, 2:1}

i = 2
nums = [fib_efficient(i, d)]
limit = 4*10**6
```

```python
while fib_efficient(i, d) < limit:
  i = i + 1
  nums.append(fib_efficient(i, d))
```

```python
nums = nums[:-1]
```

```python
evens = [i for i in nums if i % 2 == 0]
```

```python
import numpy as np
np.sum(evens) # Solution!
```

```python
# The Fibonacci sequence can also be modeled as a differenc equation:
import numpy as np
```

```python
gr   = (1 + np.sqrt(5)) / 2 # Golden Ratio
gr_c = (1 - np.sqrt(5)) / 2 # Conjugate
```

```python
def f_n(n):
    return int((gr**n - gr_c**n) / (gr - gr_c))
```

```python
f_n(0), f_n(1), f_n(2), f_n(3), f_n(4), f_n(5)
```

```python
n = np.arange(1, 50, 1)
f_n_v = np.vectorize(f_n)
y = f_n_v(n)
y = y[(y < limit) & (y % 2 == 0)] # Filtering our Fibonacci sequence with the specified conditions.
```

```python
np.sum(y)
```
