Find the sum of all the multiples of 3 or 5 below 1000.

```python
# A for-loop computes this problem when adding the right conditions:
my_sum = 0
limit = 1000
for i in range(1, limit):
  if i % 3 == 0 or i % 5 == 0:
    my_sum += i
print(my_sum)
```

```python
# However, Python can generate a list with the multiples and then we can use the sum function:
multiples = [x for x in range(1000) if x % 3 == 0 or x % 5 == 0]
print(sum(multiples))
```
