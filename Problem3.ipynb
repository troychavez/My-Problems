# Largest Prime Factor

What is the largest prime factor of 600851475143?

```python
import numpy as np
```

```python
def largest_prime_factor(num):
  prime_factors = []
  factor = 2
  while (num > 2):
    if (num % factor) == 0:
      prime_factors.append(factor)
      num = num / factor
    else:
      factor += 1
  return max(prime_factors)
```

```python
largest_prime_factor(600851475143)
```
